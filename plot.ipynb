```python
import matplotlib.pyplot as plt

n = []
normal_time = []
strassen_time = []
with open("output.txt", 'r') as f:
    next(f) 
    for line in f:
        parts = line.strip().split(',')
        n.append(int(parts[0]))
        normal_time.append(float(parts[1]))
        strassen_time.append(float(parts[2]))



plt.plot(n, normal_time, marker='o', label='Normal GEMM (O(n^3))')
plt.plot(n, strassen_time, marker='s', label='Strassen (O(n^~2.81))')
plt.xlabel('Matrix size (n x n)', fontsize=14)
plt.ylabel('Time (seconds)', fontsize=14)
plt.title(' Implementation of GEMM vs Strassen Loops', fontsize=16)
plt.legend()
plt.tight_layout()
plt.show()
```
